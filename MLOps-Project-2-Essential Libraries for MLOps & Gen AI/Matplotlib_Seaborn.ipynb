```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
x1 = np.array([1,2,3,4,5])
y1 = np.array([10,20,30,40,50])
plt.plot(x1,y1)
```

```python
plt.plot(x,y) # figure and axes will be created automatically
plt.show()
```

```python
plt.plot(x,np.sin(x), color='r', linestyle="dashed", label="sin(x)")
plt.plot(x,np.cos(x), color='blue', linestyle='-.', label="cos(x)")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

# Scatter Plot

https://matplotlib.org/stable/api/markers_api.html

```python
x = np.linspace(0,10,30)
plt.plot(x,np.sin(x), "o")
plt.plot(x,np.cos(x), "*")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
plt.scatter(x,np.sin(x))
```
